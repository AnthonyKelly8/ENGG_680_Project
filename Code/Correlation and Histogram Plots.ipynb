**Correlation for Circular**

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data from Excel file (replace 'your_file.xlsx' with your file path) for circular
data = pd.read_excel('SS CFST Database_Finalised_for correlation.xlsx', sheet_name='SS CFST Database_Circular')  # specify the sheet name if needed

# data

# Compute correlation matrix
correlation_matrix = data.corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True, cbar_kws={'shrink': .8}, annot_kws={"size": 8})
# plt.title('Rectangular Specimen')
plt.xticks(fontsize=9)  # Set x-axis label font size
plt.yticks(fontsize=9)  # Set y-axis label font size
plt.show()
```

**Histograms for Rectangular B, H**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from the Excel file (replace 'your_file.xlsx' with your file path)
file_path = 'SS CFST Database_Finalised_for correlation.xlsx'  # Replace with your Excel file path
data = pd.read_excel(file_path, sheet_name='R_CFST')  # Specify the appropriate sheet

# Define the variables to plot
variables = ['B', 'H']

# Set up the plot
plt.figure(figsize=(11, 8))

# Loop through each variable to create histograms
for i, var in enumerate(variables):
    plt.subplot(2, 2, i + 1)  # Create subplots in a 1x3 grid
    sns.histplot(data[var], bins=12,kde=True, color='darkgrey', edgecolor='black')  # Adjust bins as needed
    # plt.title(f'Histogram of {var}', fontsize=14)
    plt.xlabel(var+" (mm)", fontsize=16)
    plt.ylabel('Count', fontsize=16)
    plt.tick_params(axis='y', labelsize=13)
    plt.tick_params(axis='x', labelsize=13)

# plt.tight_layout()  # Adjust layout to prevent overlap
plt.subplots_adjust(wspace=0.25)
plt.show()
```
